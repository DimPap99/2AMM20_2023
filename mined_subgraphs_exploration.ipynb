```python
from helpers import load_pickled_data
from Graph import Graph, Subgraph
import pandas as pd
graph:Graph = load_pickled_data("working_graph3.pkl")
mined_subgraphs = load_pickled_data("Subgraphs__min12_max_None_thresh_0_2_True.pkl")
data = pd.read_csv("Results_sz__min12_max_None_thresh_0_15_True.csv")
freq = load_pickled_data("frequencies2.pkl")
for k, v in freq.items():
    
    bin_range = set()
    for bin_n, cnt in v.items():
        bin_range.add(bin_n)
    print(k, len(bin_range), bin_range)
```

```python
data['Illicit Percentage'] = data["Illicit"] / data["size"]
data['Licit Percentage'] = data["Licit"] / data["size"]
data['Unknown Percentage'] = data["Unknown"] / data["size"]
quality_group1 =  set(data[(data['Illicit Percentage'] >= 0.1) & (data['Unknown Percentage'] >= 0.4)]['Sid'].values)
quality_group2 = set(data[(data['Illicit Percentage'] < 0.1) & (data['Unknown Percentage'] < 0.2)]['Sid'].values)
```

```python
subgraphs_of_interest_1 = []
subgraphs_of_interest_2 = []

for s in mined_subgraphs:
    if s.subgraph_id in quality_group1:
        subgraphs_of_interest_1.append(s)
    if s.subgraph_id in quality_group2:
        subgraphs_of_interest_2.append(s)
```

```python
from analyse import subgraph_list
result1 = subgraph_list(graph, subgraphs_of_interest_1)
```

```python
q1_features_per_cat_no_counts = {}
for key, value in result1[1].items():
    print(key)
    print(value)
    for feat, vals in value.items():
        if feat not in q1_features_per_cat_no_counts:
            q1_features_per_cat_no_counts[feat] = set()
        for v,_ in vals.items():
            q1_features_per_cat_no_counts[feat].add(v)
```
