```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_pickled_data(file_path, verbose=False):
    try:
        data = None
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        if verbose:
            print(f"Succesfully loaded a pickled object from file {file_path}")
        return data
    except Exception as e:
        print(f"An exception occured while loading a pickled object:\n{e}")
```

```python
feat_imp = load_pickled_data('feature_importances.pkl')
```

```python
freq = load_pickled_data('frequencies.pkl')
```
